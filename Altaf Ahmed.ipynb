### Model Building

```python
x = df_encoded[['person_age',
    "person_income",
    "person_emp_exp",
    "loan_amnt",
    "credit_score"]]
y = df_encoded['loan_status']
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

## Gardio Deployment

```python
import joblib
joblib.dump(dt, 'DecisionTree.pkl')
```

```python
import gradio as gr
import pandas as pd
import joblib

# load trained model
model = joblib.load("DecisionTree.pkl")

def predict_loan(person_age, person_income, person_emp_exp, loan_amnt, credit_score):

    data = pd.DataFrame([{
        "person_age": person_age,
        "person_income": person_income,
        "person_emp_exp": person_emp_exp,
        "loan_amnt": loan_amnt,
        "credit_score": credit_score
    }])

    prediction = model.predict(data)[0]

    return "Loan Approved" if prediction == 1 else "Loan Not Approved"


interface = gr.Interface(
    fn=predict_loan,
    inputs=[
        gr.Number(label="Age"),
        gr.Number(label="Income"),
        gr.Number(label="Employment Experience (Years)"),
        gr.Number(label="Loan Amount"),
        gr.Number(label="Credit Score")
    ],
    outputs=gr.Textbox(label="Loan Status"),
    title="Loan Approval Prediction",
    description="Enter applicant details to predict loan approval"
)

interface.launch(share=True)
```
